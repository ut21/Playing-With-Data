```python
df.drop("ID NO", axis=1, inplace=False)
sns.heatmap(df.corr(), annot=True, fmt=".2f")
```

```python
df.drop(["Branch"], axis=1, inplace=True)
```

```python
df["ID NO"]=df["ID NO"].astype("str")
branch=df["ID NO"].str[0:6]
df.insert(loc=len(df.columns), column="Branch", value=branch)
df.columns
```

```python
mapping={"2022A7": "CS", "2021B5": "Repeater", "2021D2": "Repeater", "2022A1": "A1", "2022A3":"A3", "2022A4":"A4", "2022A5":"A5", "2022A8":"A8", "2022AA": "AA"}
df["Branch"]=df["Branch"].map(mapping)
```

```python
df.groupby(by="Branch")["Total (70)"].mean()
```

```python
df.groupby(by="Branch")["Total (70)"].mean().plot()
```

```python
df.groupby(by="Branch")["Total (70)"].median().plot(kind="bar", color="#96427D").set_ylabel("Median vs Branch")
```

```python
df.groupby(by="Branch")["Total (70)"].max().plot(kind="bar", color='#9D38E5', title='Max Score vs Branches').set_ylabel("Max Score")
```
